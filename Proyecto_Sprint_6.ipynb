```python
# Se importan todas las librerías que a lo largo del proyecto de usaran

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
from scipy import stats as st
import warnings
warnings.filterwarnings('ignore')
```

```python
# Visualización de dfs
# Meramente estético


def mostrar_dfs(dfs, wide_titles=None):
    """
    Muestra múltiples DataFrames con estilo HTML adaptado para modo oscuro.

    Parámetros:
    - dfs: lista de tuplas (titulo, df, head_n, [opcional] is_wide)
      ejemplo: [("Título 1", df1, 10), ("Título 2", df2, None, True)]
    - wide_titles: lista de títulos que deben mostrarse a ancho completo
    """
    if wide_titles is None:
        wide_titles = []

    html_blocks = []

    for item in dfs:
        if len(item) == 4:
            title, df, head_n, is_wide = item
        else:
            title, df, head_n = item
            is_wide = title in wide_titles

        if head_n is not None:
            df = df.head(head_n)

        table_html = df.to_html(index=True)

        css_class = "df-box wide" if is_wide else "df-box"

        block = f"""
        <div class="{css_class}">
            <h3>{title}</h3>
            {table_html}
        </div>
        """
        html_blocks.append(block)

    full_html = f"""
    <style>
        table {{
            width: 100%;
            border-collapse: collapse;
            font-size: 13px;
            background-color: #1e1e1e;
            color: #f0f0f0;
        }}
        th, td {{
            border: 1px solid #444;
            padding: 6px;
            text-align: center;
            max-width: 200px;
            white-space: nowrap;
            text-overflow: ellipsis;
            overflow: hidden;
        }}
        th {{
            background-color: #2a2a2a;
            font-weight: bold;
            color: #ffffff;
        }}
        table th:first-child,
        table td:first-child {{
            width: 1%;
        }}
        h3 {{
            margin-top: 0;
            font-size: 16px;
            color: #f2f2f2;
        }}
        .df-container {{
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(320px, 1fr));
            gap: 15px;
            margin-bottom: 30px;
        }}
        .df-box {{
            border: 1px solid #555;
            border-radius: 6px;
            padding: 10px;
            background-color: #2b2b2b;
            overflow-x: auto;
            box-shadow: 1px 1px 4px rgba(0,0,0,0.4);
        }}
        .df-box.wide {{
            grid-column: 1 / -1;
        }}
    </style>
    <div class="df-container">
        {''.join(html_blocks)}
    </div>
    """


    display(HTML(full_html))
```

```python
# Se busca verificar el comportamiento de las distribuciones de frecuencia
# al modificar sus valores NaN ya sea con cero, la media, la mediana
# o una distribución normal dada por la media y la desviación estandar de cada columna


# Sin modificar
games_copy = games.copy()

without_modificaton = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()


# Con "0"
games_copy = games.copy()

for col in ['critic_score', 'user_score', 'year_of_release']:
    games_copy[col].fillna(0, inplace=True)

with_0 = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()


# Con la media
games_copy = games.copy()

for col in ['critic_score', 'user_score', 'year_of_release']:
    games_copy[col].fillna(games_copy[col].mean(), inplace=True)

with_mean = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()


# Con la mediana
games_copy = games.copy()

for col in ['critic_score', 'user_score', 'year_of_release']:
    games_copy[col].fillna(games_copy[col].median(), inplace=True)

with_median = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()


# Con la moda
games_copy = games.copy()

for col in ['critic_score', 'user_score', 'year_of_release']:
    games_copy[col].fillna(games_copy[col].mode()[0], inplace=True)

with_mode = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()



# Con valores normalizados según la media y la desviación estandar
games_copy = games.copy()

for col in ['critic_score', 'user_score', 'year_of_release']:
    games_copy['year_of_release'] = pd.to_numeric(games['year_of_release'], errors='coerce') # Asegura 'year_of_release' sea tipo número y no haya errores con .mean y .std
    games_copy[col].fillna(np.random.normal(games_copy[col].mean(), games_copy[col].std()), inplace=True)

with_normal_dist = games_copy[['critic_score', 'user_score', 'year_of_release']].describe()



mostrar_dfs([
    ("Sin modificación", without_modificaton, 10),
    ("Modificación NaN -> 0", with_0, 10),
    ("Modificación NaN -> media", with_mean, 10),
    ("Modificación NaN -> mediana", with_median, 10),
    ("Modificación NaN -> moda", with_mode, 10),
    ("Modificación NaN -> normal dist", with_normal_dist, 10)

])

print('')
print('')
print('')
print('')


# Se realiza este proceso para poder verificar
# con que valor es más conveniente modificar los valores nulos
# visualizando como se distribuyen los datos con cada una de las agregaciones podemos tomar una decisión más acertada
columns = ['critic_score', 'user_score', 'year_of_release']

colors = {'critic_score': '#f8c8dc',
          'user_score': '#fff5ba',
          'year_of_release': 'cyan'}

games_copy = games.copy() #Es necesarfio crear esta copia ya que el games_copy se modifica desde arriba

# Método subplot devuelve el lienzo en la primera variable y los ejes de posición en la segunda, funciona así por defecto,
# enumerate() permite iterar sobre una secuencia (como una lista o tupla) y obtener tanto el índice como el valor en cada iteración.
# útil para acceder a la posición del elemento además del elemento en sí.
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(20, 20))
fig.suptitle('Distribución por métodos de imputación', fontsize=18)


for  col_idx, column in enumerate(columns):
   games_copy = games.copy() #Es necesarfio crear esta copia ya que el games_copy se modifica en los valores NaN y cambian su longitud
   normal_values = np.random.normal(
          loc=games_copy[column].mean(),
          scale=games_copy[column].std(),
          size=games_copy[column].isna().sum()
          )
   values_to_fill = {
        'sin modificar' : None,
        'cero': 0,
        'media': games_copy[column].mean(),
        'mediana': games_copy[column].median(),
        'moda': games_copy[column].mode()[0],
        'dist normal': normal_values
    }

   for row_idx, (dict_key, fill_value) in enumerate(values_to_fill.items()):
      games_copy = games.copy() # Es necesario volver a copiar el df ya que el tamaño de los NaN se modifican
      if fill_value is not None:

        # Si el método es distribución normal, se reemplazan los valores nulos con una secuencia de números
        # generados aleatoriamente según una distribución normal (media y desviación estándar de la columna).
        # se utiliza .loc para seleccionar solo las filas con NaN en esa columna y se asigna el array completo,
        # ya que tiene exactamente la misma cantidad de elementos que valores faltantes.
        if dict_key == 'dist normal':


          # Asigna 'fill_value' solo a las celdas NaN de la columna usando loc para selección condicional
          # acá es como decir a los valores NaN de las filas ingresales el valor fill_value,
          # por eso está despues, es como decir x=5, el 5 se mete en la x, el fill_value se mete en cada serie
          games_copy.loc[games_copy[column].isna(), column] = fill_value
        else:
          games_copy[column] = games_copy[column].fillna(fill_value)


      # Selecciona el subplot correspondiente y grafica el histograma de la columna imputada.
      # se personaliza el gráfico con título, nombre del eje X (nombre de la columna) y eje Y (frecuencia).
      # el título incluye el método de imputación y, si aplica, el valor usado para imputar.
      # cada ax es un gráfico en  una posición específica, luego se llena con los valores de games_copy[column]
      ax = axes[row_idx, col_idx]
      ax.hist(games_copy[column], bins=50, color=colors[column], edgecolor='black')

      # Manejo de texto para el título según el tipo de valor de imputación
      if fill_value is None:
          info = ""
      elif isinstance(fill_value, np.ndarray): # Esto quiere decir que si fill_value es un array haga lo que dice el elif


          # Aca es donde se inserta el valor promedio de los "normal_values" extrañendo la media del array generado
          # que contiene los valores de la distribución normal generada con la media y std de year_of_release
          info = f" (media fill_value dist normal: {fill_value.mean():.2f})"
      else:
          info = f" ({fill_value:.2f})"

      games_copy = games.copy() # Es necesario volver a copiar el df ya que el tamaño de los NaN se modifican
      mean_value = games_copy[column].mean()
      ax.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Media: {mean_value:.2f}')
      ax.set_title(f'Dist frec de {column} - {dict_key}{info}')
      ax.set_xlabel(column)
      ax.set_ylabel('Frecuencia')


plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Este filtrado muestra la cantidad de datos que hay por género
quant_by_genre = games.groupby('genre')['rating'].count().reset_index().rename(columns={'rating': 'count'}) # Esto es un Df porque se resetea el indice


# Obtener la moda por género

# Paso 1: Acá se extrae la moda para cada genero y se muestra el df resultante

# Ese lambda hace que 's' tome el valor de la serie de rating para cada grupo 'genre', en este caso aloja cada serie agrupada de ratings
# dada la agrupación, es decir, todos los que tengan el mismo genero se agrupan en una serie, y saca la moda de cada genero y la agrega
# a la columna 'rating'
lambda_ = lambda s: s.mode().iat[0] if not s.mode().empty else np.nan
mode_by_genre = games.groupby('genre')['rating'].agg(lambda_) # Esto es una serie


mostrar_dfs([
    ("Cantidad por género", quant_by_genre, 10),
    ("Moda por género", mode_by_genre.reset_index().rename(columns={'rating': 'mode'}), 10),
])
```

```python
# Paso 2: función de relleno
def impute(row):
    if pd.isna(row['rating']):
      return mode_by_genre[row['genre']]
    return row['rating']

games['rating'] = games.apply(impute, axis=1)

# Si quedaran NaN rezagados (géneros 100 % nulos), decide:
games['rating'].fillna('Unrated', inplace=True)
games


# Este proceso Puede ser una solución factible ya que al rellenar los valores nulos con la premisa
# de que la clasifcación de edad depende del genero de cada juego se evita sesgo hacia un valor


mostrar_dfs([
    ("Df 'games' con la moda de 'rating' en NaN", games, 10)
])
```

```python
# Se filtran los user_score válidos para Xbox One y PC
xbox_scores = games[(games['platform'] == 'XOne') & (games['user_score'].notna())]['user_score']
pc_scores = games[(games['platform'] == 'PC') & (games['user_score'].notna())]['user_score']

# Se aplica la prueba t para muestras independientes
stat1, pval1 = st.ttest_ind(xbox_scores, pc_scores, equal_var=False)
print(f"p-valor: {pval1:.4f}")
```

<div style="background-color: cyan; padding: 10px; border-radius: 5px;">
  <strong></strong>

- Si p < 0.05, se rechaza H₀ → hay diferencia significativa entre ambas plataformas.

- Si p >= 0.05, no se rechaza H₀ → no hay evidencia para decir que son distintas.



Se puede vizualizar una diferencia significtiva en las distribuciones de ambas plataformas en la clasificación de 'user_score', hay evidencia fuerte de que los usuarios califican de forma distinta en Xbox One y en PC debido a la diferencia en la experiencias de juego de ambas plataformas.

Hipótesis 2:
Comparar si las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes.

- Hipótesis nula (H₀): la media de user_score para el género Acción es igual a la de Deportes.

- Hipótesis alternativa (H₁): la media de user_score para Acción es diferente a la de Deportes.

- Tipo de prueba: t de Student para muestras independientes.

- Alfa (α): 0.05


</div>

```python
# Se filtran los user_score válidos para Action y Sports
action_scores = games[(games['genre'] == 'Action') & (games['user_score'].notna())]['user_score']
sports_scores = games[(games['genre'] == 'Sports') & (games['user_score'].notna())]['user_score']

# Se aplica la prueba t
stat2, pval2 = st.ttest_ind(action_scores, sports_scores, equal_var=False)
print(f"p-valor: {pval2:.4f}")
```

Por tanto, no se rechaza la hipótesis nula, lo que indica que no existen diferencias estadísticamente significativas entre las calificaciones de usuarios para ambos géneros; es decir, cualquier diferencia observada puede explicarse por la variabilidad natural de los datos y no por un efecto real del género sobre la puntuación.
